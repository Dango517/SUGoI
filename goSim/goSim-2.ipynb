```python
def recursive_replace(lst, old_value, new_value):
    for i, v in enumerate(lst):
        if isinstance(v, list):
            recursive_replace(v, old_value, new_value)
        else:
            if v == old_value:
                lst[i] = new_value

def realize(board_l):
    recursive_replace(board_l, 1, "し")
    recursive_replace(board_l, 2, "香")
    recursive_replace(board_l, 3, "馬")
    recursive_replace(board_l, 4, "金")
    recursive_replace(board_l, 5, "銀")
    recursive_replace(board_l, 6, "飛")
    recursive_replace(board_l, 7, "角")
    recursive_replace(board_l, 8, "王")
    recursive_replace(board_l, 9, "伏")
    recursive_replace(board_l, 0, " ")

realize(l)
```
